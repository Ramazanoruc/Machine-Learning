```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
#Verileri importluyorum

column_names=['user_id','item_id','rating','timestamp']
df=pd.read_csv('users.data',sep='\t',names=column_names)
```

```python
df
```

```python
#Diğer veriyi importluyorum(bu veride film isimleri var)
movie_titles=pd.read_csv('movie_id_titles.csv')
movie_titles
```

```python
#Bu iki veriyi birbirine merge(birlestir) ediyorum
df=pd.merge(df,movie_titles,on='item_id')
df.head()
```

```python
# Öncelikle Excel'deki pivot tablo benzeri bir yapı kuruyoruz.
# Bu yapıya göre her satır bir kullanıcı olacak şekilde (yani dataframe'imizin index'i user_id olacak)
# Sütunlarda film isimleri (yani title sütunu) olacak,
# tablo içinde de rating değerleri olacak şekilde bir dataframe oluşturuyoruz!

moviemat=df.pivot_table(index='user_id',columns='title',values='rating')
```

```python
moviemat.tail()
```

```python
type(moviemat)#Verinin tipine bakıyoruz
```

```python
#Star Wars filminin user ratinglerine bakıyorum
starwars_user_ratings=moviemat['Star Wars (1977)']
starwars_user_ratings.head()
```

#### !!!!bu uygulamada amacım kullanıcıya star wars benzeri filmleri sunmak

```python
#Corrwith metodu kullanarak star wars filmi ile korelasyonları hesaplatıyorum
#korelasyon==2 film arasındaki doğrusal bağlantı
similar_to_star_wars=moviemat.corrwith(starwars_user_ratings)
```

```python
similar_to_star_wars
```

```python
type(similar_to_star_wars) #İçinde Nan value'ler var ve bu bir seri 
#Ben bunu bir dataframe çeviricem ve içindeki NaN(bozuk,olmayan) verileri temizleyeceğim 
```

```python
#Burada hem dataframe çevirdim hem de NaN verileri sildim
corr_starwars=pd.DataFrame(similar_to_star_wars,columns=['Correlation'])
corr_starwars.dropna(inplace=True)
corr_starwars.head()
```

```python
#Şimdi elimdeki verileri korelasyona göre sıralayacağım ve bakalım star wars filmine göre en cok hangi filmi öneriyor
corr_starwars.sort_values('Correlation',ascending=False).head(10)
```

```python
#Gözüktüğü gibi sonuçlar cok alakasız bunun sebebi oylanan film sayısının azlığı
#Star wars 100 oy almıs ve baska bir film 5 oy almıs ise bu durum algoritmayı bozuyor
#sorunu düzeltmek için ratings isimli bir dataframe olusturacagım ve 100den fazla değerlendirme almıs olan filmleri ekliyeceğim
```

```python
df.head()
```

```python
#Timestamp sütünunu siliyorum,gerek yok
df.drop('timestamp',axis=1,inplace=True)
```

```python
#Her filmin ortalama rating değerini bulacağım
ratings=pd.DataFrame(df.groupby('title')['rating'].mean())
#Ardından büyükten küçüğe dogru sıralıyorum
ratings.sort_values('rating',ascending=False).head()
```

```python
#Tabikide bilmediğimiz filmler cıktı bunun sebebi ortalama alırken kaç kişi oylamıs bunu belirtmemem
#Örnek:Star Kid filmini 1 kişi oylamıstır ve 5 puan vermistir ortalaması 5 olur 
#İşte bu yüzden yazdığımız algoritmayı bozuyor bunu filtrelemem lazım
```

```python
#Burada oy sayısını hesaplıyorum
ratings['rating_oy_sayisi']=pd.DataFrame(df.groupby('title')['rating'].count())
ratings.head()
```

```python
#Burada ise büyükten küçüğe doğru sıralıyorum
#dikkat edersen -Great Day in Harlem, A (1994)- filmi 1 oy almıs ve rating'i 5 bu ve bunun gibi veriler algoritmayı bozuyor
ratings.sort_values('rating_oy_sayisi',ascending=False)
```

```python
#Rating oy sayisi sütünümuzu yapay zeka modelimize dahil ediyoruz
corr_starwars=corr_starwars.join(ratings['rating_oy_sayisi'])
```

```python
corr_starwars.head(5)
#Gözüküğü üzere modelimiz oturdu şimdi tek yapmam gereken oy sayısı 100'den büyük olanları almak
```

```python
#Filtreliyorum
corr_starwars[corr_starwars['rating_oy_sayisi']>100].sort_values('Correlation',ascending=False).head(5)
```
